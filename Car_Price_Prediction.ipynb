## Training Car Price Prediction Model Using Decision Tree Regression Algorithm

```python
predict = "price"
df = df[["symboling", "wheelbase", "carlength", "carwidth", "carheight", "curbweight", "enginesize", "boreratio", "stroke", "compressionratio", "horsepower", "peakrpm", "citympg", "highwaympg", "price"]]
x = np.array(df.drop(predict, axis=1))
y = np.array(df[predict])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
model.score(x_test, y_pred)
```

This model gives 100% accuracy on test dataset.
